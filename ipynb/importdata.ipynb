```python
# ===============================
# 1) import library ที่จำเป็น
# ===============================
import pandas as pd
from sqlalchemy import create_engine


# ===============================
# 2) ตั้งค่าการเชื่อมต่อ PostgreSQL
# ===============================
DB_USER = "myuser"
DB_PASSWORD = "mypassword"
DB_HOST = "localhost"
DB_PORT = "5432"
DB_NAME = "mydatabase"

DATABASE_URL = (
    f"postgresql+psycopg2://{DB_USER}:{DB_PASSWORD}"
    f"@{DB_HOST}:{DB_PORT}/{DB_NAME}"
)

engine = create_engine(DATABASE_URL)


# ===============================
# 3) อ่านไฟล์ CSV
# ===============================
CSV_PATH = "/Users/hilmanyusoh/Desktop/Co-opDash/data/member50.csv"

df = pd.read_csv(CSV_PATH)

# ดูข้อมูลตัวอย่าง
df.head()
```

```python
# ===============================
# 4) บันทึกข้อมูลลง PostgreSQL 
# ===============================
TABLE_NAME = "members"

# สร้างตารางใหม่ด้วย schema ที่ถูกต้อง
df.to_sql(
    TABLE_NAME,
    engine,
    if_exists="replace",   
    index=False
)

print(f"✅ Import ข้อมูลเข้า PostgreSQL สำเร็จ (สร้างตาราง '{TABLE_NAME}' ใหม่และแทนที่ข้อมูล)")
```
