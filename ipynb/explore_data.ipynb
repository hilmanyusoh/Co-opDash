```python
plt.figure()
df['gender'].value_counts().plot(kind='bar')
plt.title("Gender Count")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()

plt.figure()
df['income'].dropna().hist(bins=20)
plt.title("Income Distribution")
plt.xlabel("Income")
plt.ylabel("Count")
plt.show()

plt.figure()
df['province'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Provinces")
plt.xlabel("Province")
plt.ylabel("Count")
plt.show()

df['registration_date'] = pd.to_datetime(df['registration_date'])

plt.figure()
df['registration_date'].dt.year.value_counts().sort_index().plot()
plt.title("Members by Register Year")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```

```python
# แบ่งช่วงรายได้เป็น 5 ช่วงเท่าๆ กัน
df['income_group'] = pd.cut(df['income'], bins=5)

# นับจำนวนคนในแต่ละช่วง
group_counts = df['income_group'].value_counts().sort_index()

print("จำนวนสมาชิกตามช่วงรายได้:")
print(group_counts)
```
