```python
import psycopg2
from psycopg2 import OperationalError

# 1. กำหนดค่าตัวแปรการเชื่อมต่อ

DB_HOST = "localhost"   
DB_NAME = "mydatabase"  
DB_USER = "myuser"      
DB_PASSWORD = "mypassword" 
DB_PORT = "5432"        

def create_connection():
    """สร้างและส่งคืน Object การเชื่อมต่อ (Connection Object)"""
    connection = None
    try:
        # พยายามเชื่อมต่อฐานข้อมูล
        connection = psycopg2.connect(
            host=DB_HOST,
            port=DB_PORT,
            database=DB_NAME,
            user=DB_USER,
            password=DB_PASSWORD
        )
        print("✅ การเชื่อมต่อ PostgreSQL สำเร็จ!")
        return connection

    except OperationalError as e:
        # ดักจับข้อผิดพลาดหากเชื่อมต่อไม่สำเร็จ (เช่น Docker ไม่ได้รัน, พารามิเตอร์ผิด)
        print(f"❌ เกิดข้อผิดพลาดในการเชื่อมต่อ: '{e}'")
        print("โปรดตรวจสอบ:")
        print("1. Docker Container 'postgres' และ 'pgadmin' กำลังทำงานอยู่หรือไม่ (docker-compose ps)")
        print("2. ค่า DB_USER, DB_PASSWORD, DB_NAME ตรงกับใน docker-compose.yml หรือไม่")
        return None

# --- ทดสอบการเชื่อมต่อ ---
conn = create_connection()

# conn คือตัวแปรที่เราจะใช้ในการทำงานกับฐานข้อมูลในขั้นตอนถัดไป
```
